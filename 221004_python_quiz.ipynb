```python
# 남녀를 실행할 때마다 random하게 짝짓기를 해주는 코드를 짜시오.. 단, shuffle() 함수를 사용하시오..

from random import shuffle
l1 = ['가', '나','다', '라', '마']

shuffle(l1)  # sequence 데이터를 random 섞는 함수..
print(l1)

l2 = list(range(10))
shuffle(l2)
print(l2)
```

```python
def couple_mate(L1, L2):
    
    couples = []
    
    shuffle(L1)
    shuffle(L2)
    
    for i in range(len(L1)):
        couples.append((L1[i], L2[i]))
        
    return couples
```

```python
male = ['수퍼맨', '심봉사', '로미오', '이몽룡', '마루치']
female = ['원더우먼', '뺑덕', '줄리엣', '성춘향', '아라치']

couple_mate(male, female)
```

```python
def lotto_ball(n):
    
    balls = list(range(46))
    lotto_balls = []

    for _ in range(n):
        shuffle(balls)
        ball = balls.pop()
        lotto_balls.append(ball)

    return lotto_balls    
```
